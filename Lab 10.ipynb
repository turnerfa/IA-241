# Question 1

```python
import xlrd
book = xlrd.open_workbook('house_price.xls')
my_sheet = book.sheet_by_name("price")
price_list = []
bath_list = []
htype_list =[]
lot_size_list=[]
year_list=[]
area_list=[]

for i in range(my_sheet.nrows):
    price,bathroom,htype,lot_size,year,area = my_sheet.row_values(i) 
    if i !=0:
        (price,bathroom,htype,lot_size,year,area)
        price_list.append(price)
        bath_list.append(bathroom)
        htype_list.append(htype)
        lot_size_list.append(lot_size)
        year_list.append(year)
        area_list.append(area)
```

# Question 2

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(year_list,price_list)
plt.xlabel('YEAR')
plt.ylabel('PRICE')
plt.title('RELATIONSHIP BETWEEN PRICE AND YEAR BUILT')
plt.show()
```

# Question 3

```python
plt.hist(price_list)
plt.xlabel('PRICE')
plt.ylabel('FREQUENCY OF OCCURENCE')
plt.title('DISTRIBUTION OF HOUSE PRICES')
plt.show()
```

# Question 4

```python
point_size = [i**2.5 for i in bath_list]
plt.scatter(area_list, price_list, s=point_size)
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.title('PRICE DETERMINATION BY HOUSE AREA')
plt.show()
```
